```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ThyroidCancer.csv')
df.head()
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df = df.dropna()
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['RegionCode'] = label_encoder.fit_transform(df['Region'])
```

```python
df.loc[:, 'IodineCode'] = label_encoder.fit_transform(df['Iodine Intake'])
df
```

```python
df.loc[:, 'SeafoodCode'] = label_encoder.fit_transform(df['Seafood as main diet'])
df
```

```python
df.loc[:, 'MedicalCode'] = label_encoder.fit_transform(df['Medical'])
df
```

```python
df['Population'] = df['Population'].str.replace(',', '').astype(float)
```

```python
df['DifferenceNewCases'] = abs(df['New cases 2018'] - df['New Cases 2022'])
print(df)
```

```python
df['DifferenceDeathCases'] = abs(df['Number of Deaths 2018'] - df['Number of Deaths 2022'])
df
```

```python
region_stats = df.groupby('RegionCode')['DifferenceDeathCases'].agg(['min', 'max'])
region_stats
```

```python
region_stats['Difference'] = region_stats['max'] - region_stats['min']
region_stats['Difference']
```

```python
max_difference_region = region_stats['Difference'].idxmax()
max_difference_value = region_stats['Difference'].max()
print(f"The region with the maximum difference is: {max_difference_region}")
print(f"The maximum difference in number of deaths is: {max_difference_value}")
```

```python
region_stats1 = df.groupby('RegionCode')['DifferenceNewCases'].agg(['min', 'max'])
region_stats1
```

```python
region_stats1['Difference1'] = region_stats1['max'] - region_stats1['min']
region_stats1['Difference1']
```

```python
max_difference_region1 = region_stats1['Difference1'].idxmax()
max_difference_value1 = region_stats1['Difference1'].max()
print(f"The region with the maximum difference is: {max_difference_region1}")
print(f"The maximum difference in number of deaths is: {max_difference_value1}")
```

```python
min_difference_region = region_stats['Difference'].idxmin()
min_difference_value = region_stats['Difference'].min()
print(f"The region with the minimum difference is: {min_difference_region}")
print(f"The minimum difference in number of deaths is: {min_difference_value}")
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics 
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, classification_report
```

```python
feature_cols = ['Population', 'New Cases 2022', 'Number of Deaths 2022']
X = df[feature_cols] 
y = df['RegionCode'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 
```

```python
X_train
```

```python
X_test.shape
```

```python
y_test
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
